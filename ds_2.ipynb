```python
import matplotlib.pyplot as plt

# Sample data representing the distribution of ages and genders
ages = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70+']
male_counts = [120, 230, 350, 450, 400, 300, 200, 150]
female_counts = [110, 200, 320, 420, 380, 290, 180, 130]

# Create a figure and axis object
fig, ax = plt.subplots()

# Plotting bars for male counts
ax.bar(ages, male_counts, width=0.4, label='Male', color='skyblue')

# Plotting bars for female counts
ax.bar(ages, female_counts, width=0.4, label='Female', color='pink', bottom=male_counts)

# Adding labels and title
ax.set_xlabel('Age Group')
ax.set_ylabel('Population Count')
ax.set_title('Distribution of Ages and Genders in Population')
ax.legend()

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```
